```python
import pickle
import matplotlib.pyplot as plt 

def train_validation_comparison(file_path, title_loss, title_auc):
    with open(file_path, 'rb') as f:
        # Load the data from the pickle file
        data = pickle.load(f)

    # Extract the data
    loss = data['loss']
    val_loss = data['val_loss']
    auc = data['auc']
    val_auc = data['val_auc']

    # Create a plot with two subplots (one for the loss and the other for the AUC)
    plt.figure(figsize=(5, 8)) 

    # Plot for the Loss
    plt.subplot(2, 1, 2)
    plt.plot(loss, label='Train', color='orange')
    plt.plot(val_loss, label='Validation', color='blue')
    plt.ylim(0, 1)
    plt.title(title_loss)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # Plot for the AUC
    plt.subplot(2, 1, 1)
    plt.plot(auc, label='Train', color='orange')
    plt.plot(val_auc, label='Validation', color='blue')
    plt.ylim(0, 1)
    plt.title(title_auc)
    plt.xlabel('Epoch')
    plt.ylabel('AUC')
    plt.legend()
    plt.grid(True)

    # Display the plot
    plt.tight_layout()
    plt.show()


```

```python
import pandas as pd
import pickle

path = '../../../results_scbasset/enhancer_final/model/l2_regularization/0_5/not_filtered/visualisation_data/'

# Paths to CSV files
loss_file = f'{path}Loss_train_enhancer.csv'
auc_file = f'{path}AUC_train_enhancer.csv'
val_auc_file = f'{path}AUC_validation_enhancer.csv'
val_loss_file = f'{path}Loss_validation_enhancer.csv'

# Load data from CSV files
loss_data = pd.read_csv(loss_file)
auc_data = pd.read_csv(auc_file)
val_auc_data = pd.read_csv(val_auc_file)
val_loss_data = pd.read_csv(val_loss_file)

# Create a dictionary to store the data
combined_data = {
    'loss': loss_data['Value'].tolist(),
    'auc': auc_data['Value'].tolist(),
    'val_auc': val_auc_data['Value'].tolist(),
    'val_loss': val_loss_data['Value'].tolist()
}

# Save the dictionary in a pickle file
pickle_file = 'history.pickle'
with open(f'{path}{pickle_file}', 'wb') as f:
    pickle.dump(combined_data, f)

print("The data have been combined and saved in the pickle file:", pickle_file)
```
